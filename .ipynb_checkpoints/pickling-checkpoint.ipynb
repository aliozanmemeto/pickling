```python
# i-split dataset into 4 parts and pickle them
# ii-unlock the pickling and merge the dataset as follow 2-4-1-3 
# iii-reverse the whole data set (the last row will be the first one) 
# iV-apply sliding window technic and store the patches in a 2D list
```

```python
import pandas as pd
import numpy as np
import pickle
import statsmodels.api as sm

data = sm.datasets.co2.load_pandas().data
```

```python
data
```

```python
# i-split dataset into 4 parts and pickle them
data_parts = np.array_split(data, 4)
```

```python
for i, part in enumerate(data_parts):
    with open(f'data_part{i+1}.pkl', 'wb') as f:
        pickle.dump(part, f)
```

```python
# Load the pickled dataset parts
with open('data_part2.pkl', 'rb') as f:
    part2 = pickle.load(f)
with open('data_part4.pkl', 'rb') as f:
    part4 = pickle.load(f)
with open('data_part1.pkl', 'rb') as f:
    part1 = pickle.load(f)
with open('data_part3.pkl', 'rb') as f:
    part3 = pickle.load(f)

# Concatenate the parts in the desired order
merged_data = np.concatenate((part2, part4, part1, part3))
```

```python
merged_data
```

```python
# Reverse the order of the rows in the dataset
reversed_data = np.flip(merged_data, axis=0)
reversed_data
```

```python
source_size = 3
target_size = 1

loop_len = (len(reversed_data) - source_size)
s_list = []
t_list = []
if loop_len % target_size == 0:
    for i in range(loop_len):        
        start_ind = target_size * i
        
        s_list.append(reversed_data[start_ind : start_ind + source_size])        
        t_list.append(reversed_data[start_ind + source_size : start_ind + source_size + target_size])
              
else:
    print('Source and target size cannot be sliced')
```
